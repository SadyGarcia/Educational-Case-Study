```python
columns_heatmap_bg = ['Course',"Mother's qualification", "Father's qualification",
       "Mother's occupation", "Father's occupation"]
corr_matrix_bg = data[columns_heatmap_bg].corr()
plt.figure(figsize=(5, 5))
sns.heatmap(corr_matrix_bg, cmap='Purples', annot= True)
plt.title(' Background Correlation Matrix ')
plt.show()
```

# Linear Regression Model

Linear Regression Model is a Supervised Machine Learning Algorithm used to predict the response of a determined feature. In this case, Linear Regression models will be used to assess the academic performance based on Age and previous grades.

## Grade (1st sem) vs Grade (2nd sem)

### Build the model

```python
#Import libraries
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
```

```python
# Defining the variables
X = data[['Curricular units 1st sem (grade)']]
y = data['Curricular units 2nd sem (grade)']
```

```python
# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y,
                                                    test_size=0.2,
                                                    random_state=100)
X_train.sample(5)
```

```python
#Verify the shape
print(X_train.shape)
print(y_train.shape)
print(X_test.shape)
print(y_test.shape)
```

```python
# Define the model
LR = LinearRegression()
#Fit the model
LR.fit(X_train,y_train)
```

```python
# Printing the intercept and coefficients of the linear model
print('Intercept:',LR.intercept_)
print('Coefficient:',LR.coef_)
```

The equation gets the form :\
 Grades (2nd sem) = 0.69 + 0.896 * Grades 1st sem


### Evaluate the model

```python
# Make the predictions on the testing data
y_pred = LR.predict(X_test)
```

```python
# Calculate R-squaredd to evaluate the fitting in the model
from sklearn.metrics import r2_score
r_squared = r2_score(y_test, y_pred)
print('R-squared:', r_squared)
```

```python
# Calculate the Mean Squared Error (MSE) to evaluate the model performance
mse = mean_squared_error(y_test, y_pred)
print('Mean Squared Error:', mse)
```

```python
#Visualize the model
plt.figure(figsize=(6, 4))
plt.scatter(X_test,
            y_test,
            color='royalblue')
plt.plot(X_test,
         y_pred,
         color='red',
         linewidth=2,
         label = 'Linear Regression model ')
plt.xlabel('Grade (1st sem)')
plt.ylabel('Grade (2nd sem)')
plt.title('Linear Regression: Grade in 1st Sem vs. Grade in 2nd Sem')
plt.legend()
plt.tight_layout()
plt.show()
```

As noticed, the Linear Regression model does not have a good fit due to outliers, such as grades with a value of 0. However, the correlation is positive, indicating that if a student had good previous performance, their performance in the next semester will be similar.

### Residual Analysis

```python
#create the histofram for error terms
y_train_pred = LR.predict(X_train)
res = (y_train - y_train_pred)
plt.figure(figsize=(4,4))
sns.histplot(res,kde=True,
             color = 'skyblue')
plt.xlabel('Y_train - Y_predict')
plt.ylabel('Frequence')
plt.title('Error Terms Histogram')
plt.show()
```

```python
# Create a scatterplot for x_train and computed residual
plt.figure(figsize=(4,4))
plt.scatter(X_train,
            res,
            color='skyblue',
            label='Training Data')
plt.xlabel('Grades (1st sem)')
plt.ylabel('Error Terms')
plt.title('Distribution of Residuals')
plt.show()
```

The residual analysis has a pattern that indicates the regression model does not have a good fitting as noticed before.
